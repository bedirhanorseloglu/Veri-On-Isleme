# DEĞİŞKEN (VERİ) STANDARDİZASYONU

    Değişken dönüşümü ve değişken standardizasyonu arasındaki fark:
- Değişken Standardizasyonu -> değişkenin kendi içindeki varyans yapısı, bilgi yapısı bozulMAZ. Fakat belli bir standarda oturtulur. (Örneğin değişken olduğu şekliyle durduğunda 10 değerini düşünelim. Bu 10 değeri veri seti küçükten büyüğe sıralandığında 80. sırada olsun. Bu değişken standartlaştırıldığında bu 10 değeri muhtelemelen 1-2 gibi bir değer olacak. Yani 0 ile 1 arasında bir değer olacak. Fakat veri seti yine küçükten büyüğe sıralandığında bu defa yine 80. sırada olmuş olacak. Dolayısıyla bir değişken standarlaştırıldığında değişkenin değerleri değişecek ve belirli bir formata sokulacaktır ama taşımış olduğu yayılım mevcutta kalıyor olacak. Değişkenin ortalama, standart sapma, varyans gibi değerleri değişecektir ama taşımış olduğu bilginin değişken içindeki yayılımı değişmeyecektir. (Veri setindeki tüm sayıları aynı sayıyla çarpmak ya da bölmek))

- Değişken Dönüşümü -> dönüştürülen bir değişkenin taşıdığı bilginin taşındığı şekliyle kalamaması, dönüştürülmesi demektir. (Örneğin sex değişkenini düşünelim. Bu değişkenin değerleri kadın, erkek şeklindedir. Bunu bir dönüştürme işlemine soktuğumuzda mesela 0,1 dediğimizde yapı değişmiş olacak. Ya da 0-10 yaş arasındaki yaş değerlerine çocuk değeri atadığımızda da yine bir değişken dönüşümü uygulamış oluruz.)

Genel olarak toparlarsak;
- Değişken Standardizasyonu, veri setinin taşımış olduğu bilgiyi bozmadan belirli bir standarda getirmektir (Veri setindeki tüm sayıları aynı sayıyla çarpmak ya da bölmek). Bilgiyi değiştirir ama özünü bozmaz.
- Değişken Dönüşümü, veri setinin taşımış olduğu  bilgiyi bazen bozar bazen de temsil şeklini değiştirir.

Her bir standardizasyon aslında bir değişken dönüşümüdür. Aslında burada önemli olan mevcut değişkenin yapısı bozuluyor mu bozulmuyor mu .

```python
import pandas as pd
import numpy as np

V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame({
    "V1" : V1,
    "V2" : V2,
    "V3" : V3
})

df = df.astype(float)
df
```

- Değişken Standartlaştırma İşlemleri
    - Standardizasyon
    - Normalizasyon
    - Min-Max Dönüşümü

```python
# Standardizasyon (değerleri -3, +3 arası değerlere dönüştür. Makine öğrenmesi, algoritmalar standart değişkenleri sever.)

from sklearn import preprocessing

preprocessing.scale(df)  # df'i dönüştür dedik. -> scale()
```

```python
# Normalizasyon (değerleri 0-1 arasına dönüştürmek için kullanılır.)

from sklearn import preprocessing

preprocessing.normalize(df)     # -> normalize()
```

```python
# Min-Max Dönüşümü (istediğimiz iki aralığın arasına değerleri dönüştürmek için kullanılır.)

from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler(feature_range=(10,20))       # -> MinMaxScaler() , feature_range=(10,20) -> istediğimiz aralık
# scaler isimli bir dönüştürücü tanımladık burada aslında

scaler.fit_transform(df)
# tanımladığımız scaler isimli dönüştürücüyü çalıştırıp içine df'i attık.

# feature_range=(100,200) olsaydı değerleri on ile çarparak yazacaktı mesela. Yani dönüştürme dediğimiz şey aslında elimizdeki değerlerin hepsini aynı sayıyla genişletme ya da hepsini aynı sayıya bölme gibi düşünebiliriz.
```

- Değişken Dönüşümleri
    - 0-1 Dönüşümü
    - 1 ve Diğerleri Dönüşümü
    - Çok Sınıflı Dönüşüm

```python
import seaborn as sns 

df = sns.load_dataset("tips")
df.head()
```

```python
# 0-1 Dönüşümü (en sık kullanılan dönüşümdür.) -> LabelEncoder()

from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()
# lbe isimli bir dönüştürücü oluşturduk aslında.

lbe.fit_transform(df["sex"])    # fit_transform -> çalıştır.

# female = 0 , male = 1
```

```python
df["yeni_sex"] = lbe.fit_transform(df["sex"])
df
```

```python
# 1 ve Diğerleri Dönüşümü (Veri setinin içinde 2'den fazla sınıf varsa. Örneğin bir sınıfa 1 geri kalan sınıflara 0 değerine dönüştüreceksek.)

df.head()
```

```python
import numpy as np

df["yeni_day"] = np.where(df["day"].str.contains("Sun") , 1 , 0)        # df'in day değişkeni "Sun" ifadesi içeriyorsa 1'e, içermiyorsa 0'a dönüştürdük. 
df
```

```python
# Çok Sınıflı Dönüşüm (kategorik değişkenin sahip olduğu sınıflar sayılara dönüştürülür. Örneğin 4 sınıf varsa farklı isimleri varsa 1,2,3,4'e dönüştürür isimleri.) -> LabelEncoder

from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()
lbe.fit_transform(df["day"])    # day'in içinde 4 sınıf vardı (thur,sat,sun,fri) bunları 0,1,2,3'e dönüştürdü.
```

Burada çok dikkat etmemiz gereken bir şey var. Bu işlem sonucunda algoritma artık günleri 0,1,2,3 diye algılayacağı için (yani artık sun,thur,fri vs yok) algoritmanın kafasını karıştırabiliyor bu durum. Bunu çözmek için de "one hot dönüşümü" yapıyoruz.

- One Hot Dönüşümü ve Dummy Değişken Tuzağı

One Hot Dönüşümü, üstte dediğimiz gibi elimizdeki kategorik bir değişkenin ikiden fazla sınıfı varsa, algoritme bunun 0-3 arası olarak algıyacak. Kategorik değişkenin sınıflarının bağımlı değişkene olan etkileri bozulacaktır. Kategorik nominal iken, bozuluğ sanki aralarında bir fark varmış gibi oluyor. Halbuki bu nominal ile ölçülmüş. İşte bunun önüne geçmek için One-hat enconding yapmak gerekiyor.

```python
df.head()
```

```python
df_one_hot = pd.get_dummies(df , columns=["sex"] , prefix=["sex"])  # prefix aşağıdaki tabloda oluşan -> sex_Male	sex_Female
df_one_hot

# son iki sütun oluşmuş oldu. Burada bir satırda erkekse erkek true kadın false oldu. Alt satırda kadınsa kadın true erkek false olmuş oldu.
```

    Dummy değişken tuzağı :  
Eğer veri seti içerisindeki değişkenlere bir dönüşüm uyguladığımızda bu dönüşüm sonrasında oluşturulan yeni değişkenler birbirleri üzerinden oluşturulabiliyorsa bu durum dummy değişken tuzağı olarak adlandırılır. Ya da başka bir değişle bir değişkeni ifade eden başka bir değişken daha varsa 

Bunu çözmek için kategorik değişkenin sınıf sayısından bir az dummy değişkeni olması gerekiyor.

    Bize sağladığı iki faydası var:
- Genelde numerik değerlere dönüştürerek algoritmaları kullandığımızdan numerik değerlere dönüştürme işlemini yapmış oluyor.
- Kategorik değişkenlerin içerisindeki sınıfların etkisini veri setinde hissedilir bir hale dönüştürmemizi sağlıyor.
